```python
from bs4 import BeautifulSoup
from urllib.request import urlopen
import pandas as pd
import numpy as np
```

```python
inputFile=pd.read_excel("Input.xlsx")
inputFile
```

```python
urls=np.array(inputFile)
urls
```

```python
error=[]
for i in range(len(urls)):
    id=urls[i][0]
    v=open(f"{id}.txt","w")
for id,url in urls:
    try:
        page = urlopen(url)
        html = page.read().decode("utf-8")
        soup = BeautifulSoup(html, "html.parser")
        article_content=soup.find("div",class_="td-post-content tagdiv-type")
        article_title=soup.find("h1",class_="entry-title")
        content=article_content.get_text()
        content1=article_title.get_text()
        with open(f"{id}.txt","w") as file:
            file.write(content1)
            file.write(content)
    except:
        error.append(id)
```

```python
error2=[]
for id,url in urls:
    if id in error:
        try:
            page = urlopen(url)
            html = page.read().decode("utf-8")
            soup = BeautifulSoup(html, "html.parser")
            article_content=soup.find("div",class_="td_block_wrap tdb_single_content tdi_130 td-pb-border-top td_block_template_1 td-post-content tagdiv-type")
            article_title=soup.find("h1",class_="tdb-title-text")
            content=article_content.get_text()
            content1=article_title.get_text()
            with open(f"{id}.txt","w") as file:
                file.write(content1)
                file.write(content)
        except:
            error2.append(id)
```

```python
for id,url in urls:
    if id in error2:
        try:
            page = urlopen(url)
            html = page.read().decode("utf-8")
            soup = BeautifulSoup(html, "html.parser")
            article_content=soup.find("div",class_="td-post-content tagdiv-type")
            article_title=soup.find("h1",class_="entry-title")
            content=article_content.get_text()
            content1=article_title.get_text()
            with open(f"{id}.txt","w",encoding="utf-8") as file:
                file.write(content1)
                file.write(content)
        except:
            print(id)
```
